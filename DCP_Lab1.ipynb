```python
# Compare model-centric vs data - centric approaches uing the iris flower classification

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
x = iris.data
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
pred_dt = dt.predict(x_test)
acc_dt = accuracy_score(y_test, pred_dt)
print(acc_dt)
print("Decision Tree Accuracy:", acc_dt)

svm = SVC()
svm.fit(x_train, y_train)
pred_svm = svm.predict(x_test)
acc_svm = accuracy_score(y_test, pred_svm)
print(acc_svm)
print("SVM Accuracy:", acc_svm)

scalar = StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)

SVM_dc = SVC()
SVM_dc.fit(x_train_scaled, y_train)
pred_dc = SVM_dc.predict(x_test_scaled)
acc_dc = accuracy_score(y_test, pred_dc)
print("Data-Centric SVM Accuracy:", acc_dc)

print("\n Comparison ")
print("Decision Tree Accuracy:", acc_dt)
print("SVM Accuracy:", acc_svm)
print("Data-Centric Scalar + SVM ):", acc_dc)

import matplotlib.pyplot as plt
models = ['Decision Tree', 'SVM', 'Data-Centric SVM']
accuracies = [acc_dt, acc_svm, acc_dc]
plt.figure()
plt.bar(models, accuracies)
plt.title("Model Comparison vs Data-centric(Iris Dataset)")
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.show()
```

```python
# Recogize handwritten digit (0-9) using MNIST Dataset

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

mnist = fetch_openml('mnist_784')
x = mnist.data
y =  mnist.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
pred = knn.predict(x_test)
print("KNN Accuracy: ", accuracy_score(y_test, pred))

scalar = StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)

knn2 = KNeighborsClassifier()
knn2.fit(x_train_scaled, y_train)
pred2 = knn2.predict(x_test_scaled)
print("KNN Accuracy: ", accuracy_score(y_test, pred2))
print("Data-centric Accuracy:" , accuracy_score(y_test, pred2))

models = ["KNN", "Data-Centric KNN"]
accuracies = [accuracy_score(y_test, pred), accuracy_score(y_test, pred2)]
plt.figure()
plt.bar(models, accuracies)
plt.title("Model Comparison vs Data-centric(MNIST Dataset)")
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.show()
```
