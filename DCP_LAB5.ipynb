```python
# Data Wrangling and Transformation

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, np.nan, 30, 35, np.nan],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'Salary': [50000, 60000, np.nan, 80000, 75000],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR']
}

df = pd.DataFrame(data)

print("Original Dataset:\n", df)
print("\n-----------------------------\n")


df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].mean(), inplace=True)

print("After Missing Value Imputation:\n", df)
print("\n-----------------------------\n")


label_encoder = LabelEncoder()
df['Gender_Encoded'] = label_encoder.fit_transform(df['Gender'])

print("After Label Encoding:\n", df)
print("\n-----------------------------\n")
df = pd.get_dummies(df, columns=['Department'])

print("After Encoding:\n", df)
print("\n-----------------------------\n")


minmax_scaler = MinMaxScaler()
df[['Age_MinMax', 'Salary_MinMax']] = minmax_scaler.fit_transform(df[['Age', 'Salary']])


standard_scaler = StandardScaler()
df[['Age_Standard', 'Salary_Standard']] = standard_scaler.fit_transform(df[['Age', 'Salary']])

print("After Normalization:\n", df)
print("\n-----------------------------\n")


sales_data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Jan': [200, 150, 300],
    'Feb': [210, 160, 310],
    'Mar': [220, 170, 320]
}

sales_df = pd.DataFrame(sales_data)

print("Original Sales Data:\n", sales_df)
print("\n-----------------------------\n")


melted_df = pd.melt(sales_df, id_vars=['Name'], var_name='Month', value_name='Sales')
print("Melted Data:\n", melted_df)
print("\n-----------------------------\n")


pivot_df = melted_df.pivot(index='Name', columns='Month', values='Sales')
print("Pivoted Data:\n", pivot_df)
print("\n-----------------------------\n")


print("Final Processed Dataset:\n", df)
```

```python
# Data Wrangling and Transformation using Built-in Dataset

import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler


df = sns.load_dataset('titanic')

print("Original Dataset:\n")
print(df.head())
print("\n-----------------------------\n")


df = df[['age', 'sex', 'fare', 'class', 'embarked']]

print("Selected Columns:\n")
print(df.head())
print("\n-----------------------------\n")


df['age'].fillna(df['age'].mean(), inplace=True)


df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

print("After Missing Value Imputation:\n")
print(df.head())
print("\n-----------------------------\n")


label_encoder = LabelEncoder()
df['sex_encoded'] = label_encoder.fit_transform(df['sex'])


df = pd.get_dummies(df, columns=['class', 'embarked'])

print("After Encoding:\n")
print(df.head())
print("\n-----------------------------\n")


minmax_scaler = MinMaxScaler()
df[['age_minmax', 'fare_minmax']] = minmax_scaler.fit_transform(df[['age', 'fare']])


standard_scaler = StandardScaler()
df[['age_standard', 'fare_standard']] = standard_scaler.fit_transform(df[['age', 'fare']])

print("After Normalization:\n")
print(df.head())
print("\n-----------------------------\n")


reshaping_df = df[['age', 'fare']].head(5)


reshaping_df['Passenger'] = ['P1', 'P2', 'P3', 'P4', 'P5']

print("Before Reshaping:\n")
print(reshaping_df)
print("\n-----------------------------\n")


melted_df = pd.melt(reshaping_df, id_vars=['Passenger'], var_name='Feature', value_name='Value')

print("Melted Data:\n")
print(melted_df)
print("\n-----------------------------\n")


pivot_df = melted_df.pivot(index='Passenger', columns='Feature', values='Value')

print("Pivoted Data:\n")
print(pivot_df)
print("\n-----------------------------\n")


print("Final Processed Dataset:\n")
print(df.head())
```
